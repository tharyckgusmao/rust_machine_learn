```rust
:dep darn
:dep ndarray
:dep ndarray-csv
:dep polars = { version = "x", features = ["csv","describe","fmt","lazy","ndarray","dtype-categorical","to_dummies"]}
:dep plotters = {version = "0.3.5", features = ["evcxr", "line_series"]}
:dep linfa
:dep linfa-linear
    
extern crate csv;

use std::io::prelude::*;
use std::fs::*;
use std::path::Path
use ndarray::prelude::*;
use ndarray_csv::Array2Reader;
use ndarray::*;

use polars::prelude::*;
use std::fs::File;

use linfa::*;
use linfa::prelude::*;

use linfa::traits::{Fit, Predict};
use linfa_linear::LinearRegression;
use linfa::prelude::SingleTargetRegression;

extern crate plotters;
use plotters::prelude::*;
use plotters::series::*;
```

```rust
// import database
let df = CsvReader::from_path("./db/plano.csv").unwrap()
    .has_header(true)
    .finish().unwrap();

let df= df.clone().lazy();


let x = df.clone()
    .select([col("idade").cast(DataType::Float32)])
    .collect();

let y = df.clone().select([col("valor").cast(DataType::Float32)]).collect();


let y = y?.column("valor")?.clone();
let ndarray_x = x?.clone().to_ndarray::<Float32Type>(IndexOrder::C).unwrap();
            println!("x {:?}",ndarray_x);

// let ndarray_y =   y?.clone().to_ndarray::<Float32Type>(IndexOrder::C).unwrap();
let ndarray_y:Vec<f32> = y.f32()?.into_iter().filter_map(|opt_str| opt_str).collect();
        println!("y {:?}",ndarray_y);
let mut predict_values:Vec<f32> = vec![];
{




let ds = Dataset::new(ndarray_x.clone(), ndarray_y.clone().into());
let lin_reg = LinearRegression::new();
let model = lin_reg.fit(&ds)?;
let predict_Values = model.predict(ds);
    println!("intercept:  {}", model.intercept());
    println!("parameters: {}", model.params());
    println!("predict: {}",predict_Values.targets);

    predict_values = predict_Values.targets.to_vec();


}
    
 let figure = evcxr_figure((900, 640), |root| {
    
 let mut chart = ChartBuilder::on(&root)
        .x_label_area_size(35)
        .y_label_area_size(40)
        .margin(5)
        .caption("Scatter regression", ("sans-serif", 10.0))
        .build_cartesian_2d(0u32..90u32, 0u32..4000u32)?;  
    
         chart.configure_mesh()
                .x_label_formatter(&|x| format!("{}", *x))
                .y_label_formatter(&|y| format!("{}", *y))
                .draw()?;
        
        
         chart.draw_series(
              ndarray_x
                 .clone()
                .rows()
                .into_iter()
                .enumerate()
                .map(|(index, point)| {   
                                let x = point.clone();
                                let y = ndarray_y[index].clone() as u32;
                                let mut color = &BLUE;

                                Circle::new((x[0] as u32, y), 5, color)



                            }
                 ));

             chart.draw_series(
              ndarray_x
                 .clone()
                .rows()
                .into_iter()
                .enumerate()
                .map(|(index, point)| {   
                                let x = predict_values[index].clone() as u32;

                                let y =  ndarray_x.clone()[[index,0]]  as u32;
                                let mut color = &RED;

                                Circle::new((y, x), 5, color)



                            }
                 ));
        
                   chart.draw_series(LineSeries::new(
                        ndarray_x
                         .clone()
                        .rows()
                        .into_iter()
                        .enumerate()
                        .map(|(index, point)| {   
                                 let x = predict_values[index].clone() as u32;

                                let y =  ndarray_x.clone()[[index,0]]  as u32;
                                let mut color = &RED;

                                (y, x)
        
        
        
                                    }
                         ),
                        &RED,
                    ))?;

    
     
            Ok(())
        }).style("width:100%");

figure


     
```
